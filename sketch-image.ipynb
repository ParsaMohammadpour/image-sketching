```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img = cv2.imread('image.jpg')
cv2_imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
invert = cv2.bitwise_not(img_gray)
cv2_imshow(invert)
```

```python
blur = cv2.GaussianBlur(invert, (41, 41), 0)
cv2_imshow(blur)
```

```python
invertedblur = cv2.bitwise_not(blur)
cv2_imshow(invertedblur)
```

```python
sketch = cv2.divide(img_gray, invertedblur, scale=256.0)
cv2_imshow(sketch)
```

```python
cv2.imwrite('sketch.jpg', sketch)
```
